```python
import re
from collections import Counter
import matplotlib.pyplot as plt
from nltk import ngrams

def visualize_top_ngrams(text, n=2, top_n=20):
    # Tokenize the text into words
    text = re.sub(r'[-_/\[\](){}<>,.;!?#&@*:"%=]+', ' ', text)
    words = word_tokenize(text)
    
    # Generate n-grams from the tokenized words
    n_grams = list(ngrams(words, n))
    
    # Count the occurrences of each n-gram
    ngram_counts = Counter(n_grams)
    
    # Get the top 20 n-grams
    top_ngrams = ngram_counts.most_common(top_n)
    
    # Extract n-grams and their frequencies for visualization
    ngram_labels, ngram_frequencies = zip(*[(str(ngram), freq) for ngram, freq in top_ngrams])
    
    # Plot the top 20 n-grams
    plt.figure(figsize=(12, 6))
    plt.barh(ngram_labels, ngram_frequencies, color='skyblue')
    plt.xlabel('Frequency')
    plt.ylabel(f'{n}-grams')
    plt.title(f'Top {top_n} {n}-grams in the Text')
    plt.gca().invert_yaxis()  # Invert y-axis to display the most frequent n-grams at the top
    plt.show()

# Example usage
#text = html2text.html2text(data[0]['jobDescription'])

visualize_top_ngrams(text, n=3, top_n=20)
```

```python
def process_html(html):
    text = html2text.html2text(html)
    text = re.sub(r'[-_/\[\](){}<>,.;!?#&@*:"%=\'+\d]+', ' ', text)
    #text = remove_non_alphabetic(text)
    return text
    
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def load_vietnamese_stopwords(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        stopwords = set(line.strip() for line in file)
    return stopwords

def remove_stopwords(text,language = 'en'):
    if language == 'en':
        stop_words = set(stopwords.words('english'))
        words = nltk.word_tokenize(text)
        filtered_words = [word for word in words if word.lower() not in stop_words]
        filtered_text = ' '.join(filtered_words)
        return filtered_text
    elif language == 'vi':
        text = [word for word in text.split() if word.lower() not in vietnamese_stopwords]
        text = ' '.join(text)
        return text
    else:
        return ''
```

```python
en_text = ''
vi_text = ''
for x in data[:1]:
    temp = process_html(x['jobDescription'])
    language = detect_language(temp)
    if language == 'en':
        temp = remove_stopwords(temp,language='en')
        lemmatizer = WordNetLemmatizer()
        text = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in text]
        en_text += ' ' + temp
    else:
        temp = remove_stopwords(temp , language = 'vi')
        temp = remove_stopwords(temp, language='en')
        lemmatizer = WordNetLemmatizer()
        text = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in text]
        vi_text += ' ' + temp
```

```python
from bs4 import BeautifulSoup

html_content = data[2]['jobDescription']
soup = BeautifulSoup(html_content, 'html.parser')

requirement_keywords = ['requirement', 'skill', 'experience', 'điều kiện', 'kinh nghiệm', 'about you', 'yêu cầu bắt buộc']

requirement_tags_b = [tag for tag in soup.find_all('b') if any(keyword.lower() in tag.get_text().lower() for keyword in requirement_keywords)]
requirement_tags_p = [tag for tag in soup.find_all('p') if any(keyword.lower() in tag.get_text().lower() for keyword in requirement_keywords)]

# Kết hợp danh sách requirement_tags_b và requirement_tags_p
requirement_tags = requirement_tags_b + requirement_tags_p

# Trích xuất nội dung từ mỗi thẻ chứa từ khóa yêu cầu và lưu vào mảng
requirements = []
for tag in requirement_tags:
    # Tìm tất cả các thẻ <li> bên trong thẻ <b> hoặc <p>
    skills_li_tags = tag.find_next('ul').find_all('li', recursive=False) if tag.find_next('ul') else []
    # Trích xuất nội dung của mỗi thẻ <li> và lưu vào mảng yêu cầu
    requirements.extend([li_tag.get_text().strip() for li_tag in skills_li_tags])

# In kết quả
print("Requirements:", requirements)
```

```python
from bs4 import BeautifulSoup

html_content = data[6]['jobDescription']
soup = BeautifulSoup(html_content, 'html.parser')

# Tất cả các từ khóa mức lương cần kiểm tra
salary_keywords = ['Lương', 'Salary', 'Đãi ngộ']

# Tìm tất cả các thẻ <b> và <p> có nội dung chứa ít nhất một từ khóa mức lương
salary_tags_b = [tag for tag in soup.find_all('b') if any(keyword.lower() in tag.get_text().lower() for keyword in salary_keywords)]
salary_tags_p = [tag for tag in soup.find_all('p') if any(keyword.lower() in tag.get_text().lower() for keyword in salary_keywords)]

# Kết hợp danh sách thẻ <b> và <p>
salary_tags = salary_tags_b + salary_tags_p

# Trích xuất nội dung từ mỗi thẻ chứa từ khóa mức lương và lưu vào mảng
salaries = [tag.next_sibling.strip() for tag in salary_tags]

# In kết quả
print("Salaries:", salaries)
```
